# Diccionario de Datos: FIFA Statistics

Este diccionario describe las columnas presentes en el dataset `fifa-statistics.csv`, que contiene estadísticas de partidos de la Copa Mundial de la FIFA 2018.

| **Nombre de la Columna**         | **Descripción**                                                                 |
|----------------------------------|---------------------------------------------------------------------------------|
| `Date`                           | Fecha en que se jugó el partido (formato DD-MM-YYYY).                            |
| `Team`                           | Nombre del equipo que jugó el partido.                                           |
| `Opponent`                       | Nombre del equipo oponente.                                                     |
| `Goal Scored`                    | Número de goles anotados por el equipo.                                          |
| `Ball Possession %`              | Porcentaje de posesión de balón del equipo.                                      |
| `Attempts`                       | Número total de intentos de gol realizados por el equipo.                        |
| `On-Target`                      | Número de tiros que fueron al arco (dentro de los tres postes).                 |
| `Off-Target`                     | Número de tiros fuera de arco (fuera de los tres postes).                      |
| `Blocked`                        | Número de tiros bloqueados por la defensa del equipo oponente.                   |
| `Corners`                        | Número de tiros de esquina (corners) ganados por el equipo.                      |
| `Offsides`                       | Número de veces que el equipo fue sancionado por fuera de juego (offsides).      |
| `Free Kicks`                     | Número de tiros libres (free kicks) obtenidos por el equipo.                     |
| `Saves`                          | Número de paradas (saves) realizadas por el portero del equipo.                  |
| `Pass Accuracy %`                | Porcentaje de precisión en los pases realizados por el equipo.                   |
| `Passes`                         | Número total de pases completados por el equipo.                                 |
| `Distance Covered (Kms)`         | Distancia total recorrida por el equipo durante el partido (en kilómetros).      |
| `Fouls Committed`                | Número de faltas cometidas por el equipo.                                        |
| `Yellow Card`                    | Número de tarjetas amarillas recibidas por el equipo.                            |
| `Yellow & Red`                   | Número de tarjetas amarillas que resultaron en una tarjeta roja.                 |
| `Red`                            | Número de tarjetas rojas directas recibidas por el equipo.                       |
| `Man of the Match`               | Indica si un jugador del equipo fue elegido como el "Mejor Jugador del Partido". (`Yes`/`No`) |
| `1st Goal`                       | Minuto en el que el equipo anotó su primer gol.                                  |
| `Round`                          | Fase del torneo en la que se jugó el partido (ej. "Group Stage", "Round of 16"). |
| `PSO`                            | Indica si el partido fue decidido por penales (Penalty Shootout) (`Yes`/`No`).   |
| `Goals in PSO`                   | Número de goles anotados por el equipo en la tanda de penales (si aplica).       |
| `Own goals`                      | Número de goles en propia puerta anotados por el equipo.                         |
| `Own goal Time`                  | Minuto en el que se anotó el gol en propia puerta.                               |


# Imputación de Datos

```python
missing_values = df.isna().sum()
missing_values[missing_values > 0]
```

Tenemos 3 columnas
- 1st Goal
- Own goal
- Own goal Time

Que tienen datos incompletos, ya que son los el minuto de goles, tienen NaN los partidos que no tuvieron goles, para tratarlos los voy a reemplazar po -1 ya que no es posible una mala interpretacion

```python
# Imputamos valores faltantes
df['1st Goal'].fillna(-1, inplace= True)
df['Own goals'].fillna(-1, inplace= True)
df['Own goal Time'].fillna(-1, inplace= True)
```

```python
df.isna().sum()
```
